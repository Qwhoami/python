# Data type (자료형)
변수, 숫자, 문자열, Bool, 리스트, 튜플, 딕셔너리, 집합

```python
'''
변수
'''
a = 3
a = a+1
a
```

```python
'''
숫자
int 정수형
float 실수형
Scientific notation
2진수
8진수
16진수
'''
a =1
print(type(a))
a= 1.2
print(type(a))
a = 1e-6
print(type(a))
a = 0b011
print(a)
a = 0o110
print(a)
a = 0x10
print(a)
```

```python
'''
사칙연산
'''
a=8
b=3
print(a+b)
print(a-b)
print(a/b) # 나누기
print(a//b) # 몫
print(a%b) #나머지
print(a**b)
```

```python
'''
문자열
'''
a = " '123', '123' \n \"test\" \n"

print(a)

print(type(a))

a = '''\najskdfjas
tesetest '''
print(a)
```

```python
'''
문자열 다루기
'''
a = 'SILVACO '
b = 'python'
c = '='
print(a+b)
print(c*50)
print('length is', len(a+b))

# indexing
print(a[2])
print(a[-1])
print(a[-2])

# slicing
print(a[2:5]) # 이상 미만
print(a[2:])
print(a[1:6:2]) # 이상 미만 간격
print(a[::2])
print(a[::-1])
print(a[::-2])

# formatting
print()
```
